# Arrays Ceating with default Values


### Zeros() Function
The `zeros() function` is used to create an array filled with all zero values.
### Syntax
`
np.zeros(shape)
`
- Shape means size of array

```python
import numpy as np

Zeros_arr = np.zeros(3)
print(Zeros_arr) 
```

# Once() Function
The `ones() function` is used to create an array filled with all 1 values.
### Syntax
`
np.ones(shape)
`

```python
import numpy as np
Once = np.ones((2,3))
print(Once)
```

# Full() Function
The full() function is used to create an array where all elements are filled with a specific value given by the user.
### Syntax 
`
np.full(shape, fill_value)
`

```python
import numpy as np
Full = np.full((2,2),6)
print(Full)
```

# Check Number of Dimensions

- `ndim` is used to find the number of dimensions of an array.

```python
import numpy as np

a= np.array(22)
b= np.array([1,2,3,4,5])
c= np.array([[1,4,5],
             [5,8,9]])
d= np.array([[[0,9,7],
              [3,4,5],
              [7,6,5]]])

print(a.ndim)
print(b.ndim)
print(c.ndim)
print(d.ndim)
```
